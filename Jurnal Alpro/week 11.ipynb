```python
#soal 1 : Pemetaan Area Rawan Banjir

baris, kolom = map(int, input().split())

maks = None
minim = None
max_i = max_j = 0
min_i = min_j = 0

for i in range(baris):
    data = list(map(int, input().split()))
    
    for j in range(kolom):
        nilai = data[j]
        
        if maks is None or nilai > maks:
            maks = nilai
            max_i = i
            max_j = j
        
        if minim is None or nilai < minim:
            minim = nilai
            min_i = i
            min_j = j

print("Maksimum:", maks, "di (" + str(max_i) + ", " + str(max_j) + ")")
print("Minimum:", minim, "di (" + str(min_i) + ", " + str(min_j) + ")")
```

```python
#Soal 2: Analisis Nilai Ujian Siswa Time limit: 1 s Memory limit: 256 MB 

n = int(input())

data = list(map(int, input().split()))

maks = data[0]
minim = data[0]
total = 0

for i in range(n):
    nilai = data[i]
    
    if nilai > maks:
        maks = nilai
    
    if nilai < minim:
        minim = nilai
    
    total = total + nilai

rata = total / n

print("Nilai Maksimum:", maks)
print("Nilai Minimum:", minim)
print("Rata-rata:", round(rata, 2))
```
