```python
#soal 1

def cariangka(n1, n2, n3):
    jumlah = 0

    for i in range(1, 501):
        if str(n1) in str(i) and i % n2 != 0 and i % n3 == 0:
            print(i, end=" ")
            jumlah += 1

    print()
    print(jumlah)


n1 = int(input())
n2 = int(input())
n3 = int(input())

cariangka(n1, n2, n3)
```

```python
#soal 2

def cetaktiket(nama, film, jumlah):
    print("nama pemesan: ", nama)
    print("film yang dipilih: ", film)
    print("jumlah tiket: ", jumlah)

nama = input()
film = input()
jumlah = int(input())

cetaktiket(nama, film, jumlah)
```
