```python
def cari_area_kritis(N, target_lclu, target_hujan, data_list):
    hasil = []
    for i in range(N):
        id_obs = data_list[i][0]
        kategori = data_list[i][1]
        hujan = float(data_list[i][2])
        if kategori == target_lclu and hujan >= target_hujan:
            hasil.append(id_obs)
    return hasil

N_target = input().split()
N = int(N_target[0])
target_lclu = N_target[1]
target_hujan = float(N_target[2])

data_list = []
for i in range(N):
    data = input().split()
    data_list.append(data)

hasil = cari_area_kritis(N, target_lclu, target_hujan, data_list)

if len(hasil) > 0:
    print("Ditemukan", len(hasil), "area kritis:")
    print(" ".join(hasil))
else:
    print("Aman. Tidak ada anomali.")
```

```python
def cari_awal(arr, x):
    low = 0
    high = len(arr) - 1
    hasil = -1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == x:
            hasil = mid
            high = mid - 1
        elif arr[mid] < x:
            low = mid + 1
        else:
            high = mid - 1
    return hasil

def cari_akhir(arr, x):
    low = 0
    high = len(arr) - 1
    hasil = -1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == x:
            hasil = mid
            low = mid + 1
        elif arr[mid] < x:
            low = mid + 1
        else:
            high = mid - 1
    return hasil

def analisis_playid(N, X, arr):
    awal = cari_awal(arr, X)
    akhir = cari_akhir(arr, X)
    if awal == -1 or akhir == -1:
        print("Data Play_ID tidak ditemukan dalam pipeline.")
    else:
        total = akhir - awal + 1
        print("Target ditemukan dari indeks", awal, "sampai", akhir, ". Total frame:", total)

N_X = input().split()
N = int(N_X[0])
X = int(N_X[1])
arr = list(map(int, input().split()))

analisis_playid(N, X, arr)
```
