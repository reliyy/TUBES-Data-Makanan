```python
jumlah_tim = int(input())

data_tim = []

for i in range(jumlah_tim):
    nama, akurasi, waktu = input().split()

    tim = {
        "nama": nama,
        "akurasi": float(akurasi),
        "waktu": int(waktu)
    }

    data_tim.append(tim)

for i in range(1, len(data_tim)):

    sekarang = data_tim[i]
    posisi = i - 1

    while posisi >= 0:

        if data_tim[posisi]["akurasi"] < sekarang["akurasi"]:
            data_tim[posisi + 1] = data_tim[posisi]
            posisi -= 1

        elif (data_tim[posisi]["akurasi"] == sekarang["akurasi"] and
            data_tim[posisi]["waktu"] > sekarang["waktu"]):
            data_tim[posisi + 1] = data_tim[posisi]
            posisi -= 1

        else:
            break

    data_tim[posisi + 1] = sekarang

for tim in data_tim:
    print(f"{tim['nama']} {tim['akurasi']:.2f} {tim['waktu']}")
```

```python
n = int(input())
daftar = []
for _ in range(n):
    nama, resiko, penduduk = input().split()
    daftar.append({
        "nama": nama,
        "resiko": int(resiko),
        "penduduk": int(penduduk)
    })

for i in range(len(daftar)):
    terbaik = i
    for j in range(i + 1, len(daftar)):
        a, b = daftar[j], daftar[terbaik]
        if a["resiko"] > b["resiko"]:
            terbaik = j
        elif a["resiko"] == b["resiko"]:
            if a["penduduk"] > b["penduduk"]:
                terbaik = j
            elif a["penduduk"] == b["penduduk"] and a["nama"] < b["nama"]:
                terbaik = j
    daftar[i], daftar[terbaik] = daftar[terbaik], daftar[i]

for item in daftar:
    print(item["nama"], item["resiko"], item["penduduk"])
```
