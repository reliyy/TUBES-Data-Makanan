```python
#Soal no 3

a = int(input("masukkan umur: "))

if a < 13:
    print(a, "adalah anak - anak")
elif 13 <= a <= 17:
    print(a, "adalah seorang remaja")
elif 18 <= a <= 59:
    print(a, "adalah orang dewasa")
else:
    print(a, "adalah seorang lansia")
```

```python
#soal ke 1

jarijari = float(input("masukkan jari jari: "))
phi = 3.14

if jarijari <= 0:
    print("eror, karena jari-jari = 0")
else:
    luas = phi * (jarijari ** 2)
    keliling = 2 * phi * jarijari
    print("luas lingkaran: ", int(luas))
    print("keliling lingkaran: ", int(keliling))
```

```python
#soal ke 2 Program Aritmatika dan Perbandingan

x = float(input("masukkan angka1: "))
y = float(input("masukkan angka2: "))

tambah = x + y
kurang = x - y
kali = x * y
bagi = x / y
mod = x % y

print("penjumlahan: ", int(tambah))
print("pengurangan: ", int(kurang))
print("perkalian: ", int(kali))
print("pembagian: ", bagi)
print("modulus: ", int(mod))

if x > y:
    print("=============================")
    print("angka pertama lebih besar")
else:
    print("=============================")
    print("angka kedua lebih besar")
```

```python
#soal ke 4

x = float(input("Masukkan angka pertama: "))
y = float(input("Masukkan angka kedua: "))

if x == 0 and y == 0:
    print(x, "dan", y, "Keduanya nol")
elif x > 0 and y > 0:
    print(x, "dan", y, "Keduanya positif")
elif x < 0 or y < 0:
    print(x, "dan", y, "Salah satu negatif")
```
