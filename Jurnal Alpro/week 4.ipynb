```python
#soal 1
data = [22, 24, 26, 20, 28]

def pulihkan_data(data):
    if len(data) == 0:
        return []
    return pulihkan_data(data[1:]) + [data[0]]

data_pulih = pulihkan_data(data)
print(data_pulih)
```

```python
#soal 2
def jumlah(data):
    if len(data) == 0:
        return 0
    return data[0] + jumlah(data[1:])

def mean(data):
    return jumlah(data) / len(data)

def jumlah_selisih_kuadrat(data, rata):
    if len(data) == 0:
        return 0
    return (data[0] - rata)**2 + jumlah_selisih_kuadrat(data[1:], rata)

def variansi(data):
    rata = mean(data)
    return jumlah_selisih_kuadrat(data, rata) / len(data)

def akar(x, tebakan=1.0):
    selisih = tebakan * tebakan - x
    
    # bikin nilai mutlak manual
    if selisih < 0:
        selisih = -selisih

    if selisih < 1e-10:
        return tebakan
    
    return akar(x, (tebakan + x/tebakan) / 2)

def standar_deviasi(data):
    var = variansi(data)
    return akar(var)

print("Standar Deviasi Oksigen:", standar_deviasi(data_pulih))
```

```python
#soal 3
def filter_suhu_tinggi(data, mean):
    if len(data) == 0:
        return []
    
    if data[0] > mean:
        return [data[0]] + filter_suhu_tinggi(data[1:], mean)
    else:
        return filter_suhu_tinggi(data[1:], mean)

hasil_filter = filter_suhu_tinggi(data_pulih, 24.0)
print(hasil_filter)
```
