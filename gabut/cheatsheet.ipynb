```python
#telemetry data crisis

# 	Telemetry Data Crisis
# Description
# Di sebuah tim balap, server pengolahan data utama tiba-tiba mengalami gangguan. Sebagai seorang Data Analyst darurat, kamu diminta untuk mengolah data telemetri mentah dari mobil balap menggunakan processing unit model lama (sistem legacy).
# Sistem ini menerima data dari beberapa pembalap. Untuk setiap pembalap, dicatat nama dan waktu tempuh mereka (dalam detik) selama beberapa lap. Karena limitasi sistem legacy dan arsitektur memori yang spesifik, kepala mekanik memberikan instruksi ketat dalam pembuatan program:
# 	Data nama pembalap dan kumpulan waktu lap-nya harus dikelompokkan ke dalam satu kesatuan struktur data yang rapi.
# 	Harus ada sebuah sub-proses yang mencari dan mengembalikan nilai waktu paling cepat (fastest lap) dari keseluruhan data yang masuk.
# 	Waktu total balapan setiap pembalap harus dihitung. Namun, processing unit ini melarang keras penggunaan instruksi perulangan standar (for/while/do-while) untuk kalkulasi total waktu ini. Kalkulasi wajib dilakukan menggunakan metode di mana sebuah instruksi memanggil dirinya sendiri secara berantai hingga lap terakhir.
# 	Harus ada satu modul terpisah yang murni hanya bertugas merangkum dan mencetak hasil akhir ke layar.
# Berapakah rekor fastest lap pada sesi tersebut dan berapa total waktu masing-masing pembalap?
# Constraints
# 	1 ≤ N ≤ 100 (N adalah jumlah pembalap)
# 	1 ≤ M ≤ 100 (M adalah jumlah lap yang ditempuh)
# 	50.00 ≤ T_ij ≤ 200.00 (T_ij adalah catatan waktu pembalap ke-i pada lap ke-j, dalam bentuk bilangan desimal)
# 	Panjang karakter nama pembalap tidak lebih dari 50 karakter dan tidak mengandung spasi.
# Input
# Baris pertama berisi dua buah bilangan bulat N dan M.
# N baris berikutnya diawali dengan sebuah String yang merupakan Nama Pembalap, diikuti oleh M buah bilangan Double yang menunjukkan catatan waktu lap mereka.
# Output
# Baris pertama mencetak teks "Fastest Lap Keseluruhan: " diikuti dengan waktu tercepat (dibulatkan 2 angka di belakang koma).
# N baris berikutnya mencetak teks "[Nama Pembalap] - Total Waktu: " diikuti dengan total waktu tempuh pembalap tersebut (dibulatkan 2 angka di belakang koma).
# Sample Input 1
# Plaintext
# 3 4 Verstappen 80.52 79.21 78.80 79.05 Leclerc 81.01 80.12 79.55 80.00 Hamilton 80.20 79.88 79.91 79.44
# Sample Output 1
# Plaintext
# Fastest Lap Keseluruhan: 78.80 Verstappen - Total Waktu: 317.58 Leclerc - Total Waktu: 320.68 Hamilton - Total Waktu: 319.43
# Sample Input 2
# Plaintext
# 2 3 Alonso 85.10 84.55 84.22 Norris 84.90 84.00 84.60
# Sample Output 2
# Plaintext
# Fastest Lap Keseluruhan: 84.00 Alonso - Total Waktu: 253.87 Norris - Total Waktu: 253.50


class Pembalap:
    def __init__(self, nama, lap):
        self.nama = nama
        self.lap = lap


def total_waktu(arr, idx):
    if idx == len(arr):
        return 0

    return arr[idx] + total_waktu(arr, idx + 1)


def fastest_lap(data):
    fastest = data[0].lap[0]

    for pembalap in data:
        for waktu in pembalap.lap:
            if waktu < fastest:
                fastest = waktu

    return fastest


def print_hasil(data, fastest):
    print(f"Fastest Lap Keseluruhan: {fastest:.2f}")

    for pembalap in data:
        total = total_waktu(pembalap.lap, 0)

        print(f"{pembalap.nama} - Total Waktu: {total:.2f}")


n, m = map(int, input().split())

data = []

for _ in range(n):
    temp = input().split()

    nama = temp[0]
    lap = list(map(float, temp[1:]))

    data.append(Pembalap(nama, lap))

fastest = fastest_lap(data)

print_hasil(data, fastest)
```

```python
#Panen Koperasi

# 	Digitalisasi Panen Koperasi
# Description
# Sebuah Koperasi Unit Desa (KUD) sedang melakukan transformasi digitalisasi pertanian. Mereka mengumpulkan data rekapitulasi hasil panen dari para petani setiap periodenya. Sayangnya, modul penghitungan pada aplikasi KUD belum selesai sepenuhnya, sehingga mereka membutuhkan skrip tambahan sementara.
# Program darurat ini harus memproses nama petani dan riwayat berat panen harian mereka (dalam satuan kg). KUD menerapkan beberapa syarat ketat dalam pembuatan skrip ini:
# 	Riwayat panen masing-masing petani beserta nama mereka harus disimpan dalam sebuah struktur data yang saling terikat.
# 	Program harus bisa menghitung rata-rata hasil panen keseluruhan desa (gabungan dari semua petani). Proses kalkulasi rata-rata ini harus dikembalikan oleh sebuah sub-program khusus.
# 	KUD ingin memberikan insentif kepada petani yang performanya stabil. Oleh karena itu, program harus bisa menghitung berapa hari seorang petani berhasil mendapatkan hasil panen di atas batas target (Threshold) yang ditentukan.
# 	Penting: Pihak KUD melarang penggunaan perulangan standar (seperti for atau while) untuk mengecek kondisi di atas batas target tersebut. Kamu harus membangun modul yang memanggil dirinya sendiri secara berantai untuk melakukan pengecekan satu per satu hingga data harian habis.
# 	Sediakan satu blok sub-program terpisah yang tidak mengembalikan nilai apa-apa, murni hanya bertugas mencetak laporan akhir sesuai format yang diminta.
# Berapa rata-rata panen desa tersebut, dan berapa hari masing-masing petani menembus target panen?
# Constraints
# 	1 ≤ N ≤ 100 (N adalah jumlah petani)
# 	1 $\le$ $M$ $\le$ 30 (M adalah rentang hari panen yang dicatat)
# 	10.0 ≤ P_ij ≤ 500.0 (P_ij adalah hasil panen harian dalam kg, berupa bilangan desimal)
# 	10.0 ≤ T ≤ 500.0 (T adalah batas target panen harian untuk mendapatkan insentif)
# 	Panjang karakter nama petani tidak lebih dari 50 karakter dan tidak mengandung spasi.
# Input
# Baris pertama berisi tiga buah bilangan: N, M, dan T.
# N baris berikutnya diawali dengan sebuah kata yang merupakan Nama Petani, diikuti oleh M buah bilangan desimal yang menunjukkan catatan panen mereka.
# Output
# Baris pertama mencetak teks "Rata-rata Panen Desa: " diikuti dengan nilai rata-rata keseluruhan (dibulatkan 2 angka di belakang koma).
# $N$ baris berikutnya mencetak teks "[Nama Petani] tembus target sebanyak [X] hari".
# Sample Input 1
# 3 4 50.0
# Budi 45.5 55.0 60.5 48.0
# Siti 51.0 52.0 49.5 50.1
# Joko 40.0 42.5 45.0 48.5
# Sample Output 1
# Rata-rata Panen Desa: 48.97
# Budi tembus target sebanyak 2 hari
# Siti tembus target sebanyak 3 hari
# Joko tembus target sebanyak 0 hari
# Sample Input 2
# 2 3 100.0
# Agus 105.0 95.5 110.0
# Dewi 99.0 100.5 102.0
# Sample Output 2
# Rata-rata Panen Desa: 102.00
# Agus tembus target sebanyak 2 hari
# Dewi tembus target sebanyak 2 hari


class Petani:
    def __init__(self, nama, panen):
        self.nama = nama
        self.panen = panen


def hitung_target(arr, idx, target):
    if idx == len(arr):
        return 0

    if arr[idx] > target:
        return 1 + hitung_target(arr, idx + 1, target)

    return hitung_target(arr, idx + 1, target)


def rata_desa(data):
    total = 0
    jumlah = 0

    for petani in data:
        for hasil in petani.panen:
            total += hasil
            jumlah += 1

    return total / jumlah


def print_hasil(data, target, rata):
    print(f"Rata-rata Panen Desa: {rata:.2f}")

    for petani in data:
        jumlah = hitung_target(petani.panen, 0, target)

        print(f"{petani.nama} tembus target sebanyak {jumlah} hari")


n, m, target = input().split()

n = int(n)
m = int(m)
target = float(target)

data = []

for _ in range(n):
    temp = input().split()

    nama = temp[0]
    panen = list(map(float, temp[1:]))

    data.append(Petani(nama, panen))

rata = rata_desa(data)

print_hasil(data, target, rata)
```

```python
#Algoritma Bullish Stock

# Description
# Sebuah firma investasi sedang melakukan analisis risiko dan pemantauan tren dari berbagai aset saham. Server analitik utama sedang dalam perbaikan, sehingga seorang Data Analyst harus menulis skrip sementara untuk mengevaluasi data pergerakan harga penutupan saham harian.
# Data yang masuk berupa nama kode saham (ticker) dan riwayat harganya selama beberapa hari. Manajer investasi meminta skrip ini memenuhi kriteria struktural yang ketat:
# 1.	Kode saham dan riwayat harganya harus dikemas dalam satu tipe bentukan yang terstruktur.
# 2.	Dibutuhkan sebuah fungsi untuk mencari harga tertinggi (All-Time High) dari seluruh data saham yang masuk selama periode tersebut.
# 3.	Dibutuhkan evaluasi tren "BULLISH". Sebuah saham dikatakan bullish apabila harganya dari hari ke hari selalu naik atau bertahan (tidak pernah turun sama sekali dibandingkan hari sebelumnya). Evaluasi ini wajib dilakukan menggunakan sub-program yang memanggil dirinya sendiri secara berantai. Pengecekan tren ini tidak boleh menggunakan looping standar.
# 4.	Sediakan satu prosedur terpisah untuk mencetak ringkasan hasil analisis ke layar.

# Berapakah rekor harga tertinggi secara keseluruhan, dan manakah saham yang memiliki tren pergerakan bullish?
# Constraints
# 	1 ≤ N ≤ 100 ($N$ adalah jumlah saham)
# 	2 ≤ M$ $≤ 30 ($M$ adalah jumlah hari pencatatan)
# 	1.0 ≤ H_ij 10000.0 ($H_{ij}$ adalah harga saham ke-i pada hari ke-j, berupa bilangan desimal)
# 	Panjang kode saham maksimal 10 karakter huruf kapital tanpa spasi.
# Input
# Baris pertama berisi dua buah bilangan bulat $N$ dan $M$.
# $N$ baris berikutnya diawali dengan sebuah String yang merupakan Ticker Saham, diikuti oleh $M$ buah bilangan Double yang menunjukkan harga harian.
# Output
# Baris pertama mencetak teks "All-Time High Keseluruhan: " diikuti dengan harga tertinggi (dibulatkan 2 angka di belakang koma).
# $N$ baris berikutnya mencetak teks "[Ticker] - Tren: [BULLISH / NON-BULLISH]".

# Sample Input 1
# 3 4
# BBCA 8000.0 8100.0 8150.0 8200.0
# GOTO 150.0 140.0 145.0 130.0
# TLKM 3000.0 3000.0 3100.0 3150.0
# Sample Output 1
# All-Time High Keseluruhan: 8200.00
# BBCA - Tren: BULLISH
# GOTO - Tren: NON-BULLISH
# TLKM - Tren: BULLISH
# Sample Input 2
# 2 3
# AMMN 1000.5 900.0 1200.0
# BREN 5500.0 6000.0 6000.0
# Sample Output 2
# All-Time High Keseluruhan: 6000.00
# AMMN - Tren: NON-BULLISH
# BREN - Tren: BULLISH




class Saham:
    def __init__(self, kode, harga):
        self.kode = kode
        self.harga = harga


def bullish(arr, idx):
    if idx == len(arr) - 1:
        return True

    if arr[idx] > arr[idx + 1]:
        return False

    return bullish(arr, idx + 1)


def all_time_high(data):
    maks = data[0].harga[0]

    for saham in data:
        for harga in saham.harga:
            if harga > maks:
                maks = harga

    return maks


def print_hasil(data, ath):
    print(f"All-Time High Keseluruhan: {ath:.2f}")

    for saham in data:
        if bullish(saham.harga, 0):
            tren = "BULLISH"
        else:
            tren = "NON-BULLISH"

        print(f"{saham.kode} - Tren: {tren}")


n, m = map(int, input().split())

data = []

for _ in range(n):
    temp = input().split()

    kode = temp[0]
    harga = list(map(float, temp[1:]))

    data.append(Saham(kode, harga))

ath = all_time_high(data)

print_hasil(data, ath)
```

```python
#Rekapitulasi Skor Lomba



class Tim:
    def __init__(self, nama, skor):
        self.nama = nama
        self.skor = skor


def best_score(arr, idx):
    if idx == len(arr) - 1:
        return arr[idx]

    next_score = best_score(arr, idx + 1)

    if arr[idx] > next_score:
        return arr[idx]

    return next_score


def rata_global(data):
    total = 0
    jumlah = 0

    for tim in data:
        for nilai in tim.skor:
            total += nilai
            jumlah += 1

    return total / jumlah


def print_hasil(data, rata):
    print(f"Rata-rata Global: {rata:.2f}")

    for tim in data:
        best = best_score(tim.skor, 0)

        print(f"{tim.nama} - Best Score: {best:.2f}")


n, m = map(int, input().split())

data = []

for _ in range(n):
    temp = input().split()

    nama = temp[0]
    skor = list(map(float, temp[1:]))

    data.append(Tim(nama, skor))

rata = rata_global(data)

print_hasil(data, rata)
```

```python
#Audio Streaming Analytics



class User:
    def __init__(self, nama, durasi):
        self.nama = nama
        self.durasi = durasi


def cari_hari(arr, idx, target):
    if idx == len(arr):
        return -1

    if arr[idx] > target:
        return idx + 1

    return cari_hari(arr, idx + 1, target)


def total_server(data):
    total = 0

    for user in data:
        for durasi in user.durasi:
            total += durasi

    return total


def print_hasil(data, target, total):
    print(f"Total Durasi Server: {total:.2f} menit")

    for user in data:
        hari = cari_hari(user.durasi, 0, target)

        if hari == -1:
            print(f"{user.nama} belum mencapai target durasi")
        else:
            print(f"{user.nama} menembus target pertama kali pada hari ke-{hari}")


n, m, target = input().split()

n = int(n)
m = int(m)
target = float(target)

data = []

for _ in range(n):
    temp = input().split()

    nama = temp[0]
    durasi = list(map(float, temp[1:]))

    data.append(User(nama, durasi))

total = total_server(data)

print_hasil(data, target, total)
```
