```python
class Mahasiswa:
    def __init__(self, nama, nilai):
        self.nama = nama
        self.nilai = nilai

    def tampil(self):
        print(self.nama, self.nilai)

m1 = Mahasiswa("Rayan", 90)
m1.tampil()
```
