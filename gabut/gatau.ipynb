```python
class Robot:
    def __init__(self, robot_id, tekanan):
        self.robot_id = robot_id
        self.tekanan = tekanan


def tekanan_maksimum(data):
    maksimum = data[0].tekanan[0]

    for robot in data:
        for nilai in robot.tekanan:
            if nilai > maksimum:
                maksimum = nilai

    return maksimum


def tekanan_minimum(arr, idx):
    if idx == len(arr) - 1:
        return arr[idx]

    minimum_selanjutnya = tekanan_minimum(arr, idx + 1)

    if arr[idx] < minimum_selanjutnya:
        return arr[idx]

    return minimum_selanjutnya


def print_hasil(data, maksimum):
    print(f"Tekanan Maksimum Global: {maksimum:.2f}")

    for robot in data:
        minimum = tekanan_minimum(robot.tekanan, 0)

        print(f"Probe {robot.robot_id} - Tekanan Minimum: {minimum:.2f}")


n, m = map(int, input().split())

data = []

for _ in range(n):
    temp = input().split()

    robot_id = temp[0]
    tekanan = list(map(float, temp[1:]))

    data.append(Robot(robot_id, tekanan))

maksimum = tekanan_maksimum(data)

print_hasil(data, maksimum)
```

```python
class Batch:
    def __init__(self, batch_id, skor):
        self.batch_id = batch_id
        self.skor = skor


def persentase_global(data, batas):
    lolos = 0
    total = 0

    for batch in data:
        for nilai in batch.skor:
            total += 1

            if nilai >= batas:
                lolos += 1

    return (lolos / total) * 100


def cek_batch(arr, idx, batas):
    if idx == len(arr):
        return True

    if arr[idx] < batas:
        return False

    return cek_batch(arr, idx + 1, batas)


def print_hasil(data, batas, persen):
    print(f"Persentase Kelolosan Global: {persen:.2f}%")

    for batch in data:
        if cek_batch(batch.skor, 0, batas):
            status = "FLAWLESS"
        else:
            status = "DEFECTIVE"

        print(f"Batch {batch.batch_id} - Status: {status}")


n, m, batas = input().split()

n = int(n)
m = int(m)
batas = float(batas)

data = []

for _ in range(n):
    temp = input().split()

    batch_id = temp[0]
    skor = list(map(float, temp[1:]))

    data.append(Batch(batch_id, skor))

persen = persentase_global(data, batas)

print_hasil(data, batas, persen)
```

```python
class Truk:
    def __init__(self, plat, bbm):
        self.plat = plat
        self.bbm = bbm


def total_bbm(arr, idx):
    if idx == len(arr):
        return 0

    return arr[idx] + total_bbm(arr, idx + 1)


def cari_terendah(data):
    minimum = total_bbm(data[0].bbm, 0)

    for truk in data:
        total = total_bbm(truk.bbm, 0)

        if total < minimum:
            minimum = total

    return minimum


def hitung_boros(arr, idx, batas):
    if idx == len(arr):
        return 0

    if arr[idx] > batas:
        return 1 + hitung_boros(arr, idx + 1, batas)

    return hitung_boros(arr, idx + 1, batas)


def print_hasil(data, batas, minimum):
    print(f"Total BBM Terendah: {minimum:.2f}")

    for truk in data:
        boros = hitung_boros(truk.bbm, 0, batas)

        print(f"{truk.plat} - Hari Boros: {boros} hari")


n, m, batas = input().split()

n = int(n)
m = int(m)
batas = float(batas)

data = []

for _ in range(n):
    temp = input().split()

    plat = temp[0]
    bbm = list(map(float, temp[1:]))

    data.append(Truk(plat, bbm))

minimum = cari_terendah(data)

print_hasil(data, batas, minimum)
```

```python
class User:
    def __init__(self, nama, halaman):
        self.nama = nama
        self.halaman = halaman


def total_platform(data):
    total = 0

    for user in data:
        for nilai in user.halaman:
            total += nilai

    return total


def gagal_pertama(arr, idx, target):
    if idx == len(arr):
        return 0

    if arr[idx] < target:
        return idx + 1

    return gagal_pertama(arr, idx + 1, target)


def print_hasil(data, target, total):
    print(f"Total Literasi Platform: {total:.2f} Halaman")

    for user in data:
        gagal = gagal_pertama(user.halaman, 0, target)

        print(f"{user.nama} - Gagal pada hari ke: {gagal}")


n, m, target = input().split()

n = int(n)
m = int(m)
target = float(target)

data = []

for _ in range(n):
    temp = input().split()

    nama = temp[0]
    halaman = list(map(float, temp[1:]))

    data.append(User(nama, halaman))

total = total_platform(data)

print_hasil(data, target, total)
```
