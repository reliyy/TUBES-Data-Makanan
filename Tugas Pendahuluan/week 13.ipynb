```python
def linear_search(data, target):
    for i in range(len(data)):
        if data[i] == target:
            return i
    return -1

n, target = map(int, input().split())
nim = list(map(int, input().split()))

hasil = linear_search(nim, target)

if hasil != -1:
    print(f"Ditemukan pada indeks {hasil}")
else:
    print("Belum Mengumpulkan")
```

```python
def binary_search(data, target):
    kiri = 0
    kanan = len(data) - 1

    while kiri <= kanan:
        tengah = (kiri + kanan) // 2

        if data[tengah] == target:
            return tengah

        elif target < data[tengah]:
            kanan = tengah - 1

        else:
            kiri = tengah + 1

    return -1


n, target = map(int, input().split())
tiket = list(map(int, input().split()))

hasil = binary_search(tiket, target)

if hasil != -1:
    print(f"Valid! Tiket ditemukan pada indeks {hasil}")
else:
    print("Tiket Palsu")
```
