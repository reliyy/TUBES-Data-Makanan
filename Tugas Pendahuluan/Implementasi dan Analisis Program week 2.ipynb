```python
#Sistem Parkir Kampus

lama_parkir = int(input("masukkan lama parkir: "))
biaya_parkir = lama_parkir * 2000

if lama_parkir > 5:
    total_parkir = biaya_parkir + 5000
else:
    total_parkir = biaya_parkir


status_member = input("member: ")
if status_member == "true" or status_member == "True":
    x = total_parkir * 0.10
    hasil_akhir = total_parkir - x
else:
    hasil_akhir = total_parkir

print("Total Biaya Parkir: Rp ", int(hasil_akhir))
```

```python
#Deteksi Angka Spesial

angka = float(input("masukkan angka: "))

if 100 >= angka > 0:
    print(angka, "adalah bilangan positif")
elif angka > 100:
    print(angka, "adalah bilangan positif dan lebih dari 100")
elif angka < 0:
    print(angka, "adalah bilangan negatif")
else:
    print(angka, "bilangan tersebut nol")
```

```python
#Sistem diskon toko buku

belanja = int(input("masukkan total belanja: Rp "))

if belanja > 200000:
    x = belanja * 0.15
    total_belanja = belanja - x
    print("belanja mendapat diskon 15%, jadi Rp", int(total_belanja))
elif belanja > 100000:
    y = belanja * 0.10
    total_belanja = belanja -y
    print("belanja mendapat diskon 10%, jadi Rp ", int(total_belanja))
else:
    total_belanja = belanja
    print("total belanja Rp ", total_belanja)

pembeli = input("mahasigma atau bukan? ")

if pembeli == "true" or pembeli == "True":
    a = total_belanja * 0.05
    diskon_akhir = total_belanja - a
    print("Karena pembeli adalah mahasigma mendapat diskon 5%, jadi Rp ", int(diskon_akhir))
else:
    diskon_akhir = total_belanja
```
