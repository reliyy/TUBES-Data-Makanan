```python
armada = []  

def daftar_pesawat():
    if len(armada) >= 50:
        print("Kapasitas armada sudah penuh!")
        return

    print("\n=== Input Data Pesawat ===")
    nama = input("Nama pesawat: ")
    kargo = int(input("Kapasitas kargo: "))
    jarak = int(input("Jarak tempuh: "))

    data = {
        "nama": nama,
        "kargo": kargo,
        "jarak": jarak
    }

    armada.append(data)
    print("Pesawat berhasil ditambahkan!\n")

def hitung_energi(jarak, kargo):
    energi = (2 * jarak) + kargo
    return energi

def kalibrasi(warp):
    if warp == 1:
        return 10
    else:
        return (2 * kalibrasi(warp - 1)) + 5

print("=== SISTEM ARMADA GALACTICA VANGUARD ===")

jumlah = int(input("Masukkan jumlah pesawat (min 2): "))

while jumlah < 2:
    jumlah = int(input("Minimal 2 pesawat, ulangi: "))

for i in range(jumlah):
    print(f"\nPesawat ke-{i+1}")
    daftar_pesawat()

print("\n=== LAPORAN MANIFES PENERBANGAN ===")

for p in armada:
    energi = hitung_energi(p["jarak"], p["kargo"])
    print(f"Pesawat: {p['nama']}")
    print(f"Energi dibutuhkan: {energi}")
    print("-" * 25)

warp = int(input("\nMasukkan tingkat warp hari ini: "))
kode = kalibrasi(warp)

print(f"Kode kalibrasi hyperdrive: {kode}")
print("Armada siap diterbangkan")
```
