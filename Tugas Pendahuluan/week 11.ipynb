```python
#Soal : Analisis Sensor Suhu

n = int(input())

data = list(map(float, input().split()))

tinggi = data[0]
rendah = data[0]

for i in range(n):
    nilai = data[i]

    if nilai > tinggi:
        tinggi = nilai
    
    if nilai < rendah:
        rendah = nilai

selisih = tinggi - rendah

tinggi = round(tinggi, 2)
rendah = round(rendah, 2)
selisih = round(selisih, 2)

tinggi_str = str(tinggi)
rendah_str = str(rendah)
selisih_str = str(selisih)

if "." in tinggi_str:
    if len(tinggi_str.split(".")[1]) == 1:
        tinggi_str = tinggi_str + "0"
else:
    tinggi_str = tinggi_str + ".00"

if "." in rendah_str:
    if len(rendah_str.split(".")[1]) == 1:
        rendah_str = rendah_str + "0"
else:
    rendah_str = rendah_str + ".00"

if "." in selisih_str:
    if len(selisih_str.split(".")[1]) == 1:
        selisih_str = selisih_str + "0"
else:
    selisih_str = selisih_str + ".00"

print(tinggi_str, rendah_str, selisih_str)
```
