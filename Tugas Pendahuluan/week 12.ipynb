```python
n = int(input())
nilai = list(map(int, input().split()))

for i in range(1, n):
    x = nilai[i]
    j = i - 1

    while j >= 0 and nilai[j] > x:
        nilai[j + 1] = nilai[j]
        j -= 1

    nilai[j + 1] = x

for i in range(n):
    print(nilai[i], end=" ")
```

```python
n = int(input())
buku = list(map(int, input().split()))

for i in range(n):
    min_idx = i

    for j in range(i + 1, n):
        if buku[j] < buku[min_idx]:
            min_idx = j

    buku[i], buku[min_idx] = buku[min_idx], buku[i]

for i in range(n):
    print(buku[i], end=" ")
```
