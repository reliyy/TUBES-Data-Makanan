```python
data = [22, 24, 26, 20, 28]

def jumlah(data, i=0):
    if i == len(data):
        return 0
    return data[i] + jumlah(data, i + 1)

def rata_rata(data):
    total = jumlah(data)
    return total / len(data)

def totalSelisihKuadrat(data, mean, i=0):
    if i == len(data):
        return 0
    return (data[i] - mean) ** 2 + totalSelisihKuadrat(data, mean, i + 1)

def variansi(data):
    mean = rata_rata(data)
    total_selisih = totalSelisihKuadrat(data, mean)
    return total_selisih / len(data)

total = jumlah(data)
mean = rata_rata(data)
total_selisih = totalSelisihKuadrat(data, mean)
var = variansi(data)

print("Total Massa Suhu:", total)
print("Rata-rata Suhu:", mean)
print("Total Selisih Kuadrat:", total_selisih)
print("Variansi Akhir:", var)
```
