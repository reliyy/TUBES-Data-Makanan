```python
#soal ke 1

#fungsi untuk mengecek apakah tahun itu tahun kabisat
def tahunkabisat(tahun):

    #kondisi pertama : jika tahun habis dibagi 400, berarti tahun itu kabisat
    if tahun % 400 == 0:
        return True
    
    #kondisi kedua : jika tahun habis dibagi 100, berarti bukan tahun kabisat
    elif tahun % 100 == 0:
        return False
    
    #kondisi ketiga : jika tahun habis dibagi 4, berarti tahun kabisat
    elif tahun % 4 == 0:
        return True
    
    #kondisi keempat : jika semua kondisi tidak terpenuhi maka bukan tahun kabisat
    else:
        return False

#meminta masukan
tahun = int(input("masukkan tahun: "))
#mencetak hasil
print(tahunkabisat(tahun))
```

```python
#soal ke 2

#fungsi untuk menghitung angka dan pangkat sesuai posisi angkanya
def jumlahpangkat(n):
    
    #untuk menampung hasil penjumlahan
    total = 0
    
    #mengubah ke string agar dapat dihitung per karakter yg dimasukkan
    n_str = str(n)

    #membuat perulangan sebanyak jumlah yang diinput n
    for i in range(len(n_str)):
        
        #mengambil digit ke-i lalu diubah ke integer agar dapat dihitung
        digit = int(n_str[i])
        #pangkatkan digit dengan posisinya dimulai dari 1
        total += digit ** (i + 1)

    #mengembalikan hasil jumlah
    return total

#meminta masukan
n = int(input("Masukkan angka: "))
#mencetak hasil
print(jumlahpangkat(n))
```

```python
#soal ke 3

#fungsi untuk membalik urutan bilangan
def reversenumber(n):
    #sebagai penampung hasil reverse
    x = 0
    #hitung jumlah bilangan dari n dan mengubahnya ke string agar dapat terhitung
    angka = len(str(n))

    #melakukan perulangan sebanyak jumlah n
    for i in range(angka):
        #ambil digit terakhir dari n menggunakan mod 10
        jumlah = n % 10
        #untuk menggeser digit yang ada di kiri ke satu posisi dan menaruh digit baru di paling kanan
        x = x * 10 + jumlah
        #Hapus digit terakhir dari n dengan pembagian bulat 10
        n = n // 10
    
    #kembalikan hasil reverse
    return x

#meminta masukan
n = int(input())
#mencetak hasil
print(reversenumber(n))
```

```python
#soal ke 4

#membuat fungsi untuk mengecek angka spesial
def angkaspesial(n):
    #digunakan sebagai penampung hasil penjumlahan
    jumlah = 0
    #digunakan sebagai penampung banyaknya digit
    total = 0

    #perulangan setiap n yang sudah diubah ke string
    for i in str(n):
        #tambahkan nilai digit ke jumlah
        jumlah += int(i)
        #menghitung banyak digit
        total += 1

    #membuat kondisi jika digit dikali banyak digit apakah sama dengan n
    if jumlah * total == n:
        return True
    else:
        return False

#meminta masukan
n = int(input())
#mencetak hasil
print(angkaspesial(n))
```

```python
#soal ke 5

#fungsi untuk menghitung jumlah digit bergantian
def jumlahAngka(n):
    #untuk menampung jumlah akhir
    hasil = 0
    #sebagai penentu angka +/- (dimulai dari 1 (positif))
    tanda = 1

    #membuat perulangan digit dari n yang sudah di ubah ke string
    for i in str(n):
        #mengalikan digit dengan tanda lalu tambahkan ke hasil
        hasil += tanda * int(i)
        #balik tanda untuk digit berikutnya
        tanda *= -1

    #mengembalikan hasil penjumlahan
    return hasil

#meminta masukan
n = int(input())
#mencetak hasil    
print(jumlahAngka(n))
```
